```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

```

```python
# Read image in grayscale mode (8-bit)
img = cv2.imread('cat.54.jpg', cv2.IMREAD_GRAYSCALE)

# Check if image is loaded
if img is None:
    print("Image not found!")
else:
    print("Image shape:", img.shape)
    print("Image dtype:", img.dtype)  # Should be uint8
```

```python
plt.figure(figsize=(6,6))
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
hist_eq = cv2.equalizeHist(img)

plt.figure(figsize=(6,6))
plt.imshow(hist_eq, cmap='gray')
plt.title("Histogram Equalized")
plt.axis('off')
plt.show()
```

```python
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
clahe_img = clahe.apply(img)

plt.figure(figsize=(6,6))
plt.imshow(clahe_img, cmap='gray')
plt.title("CLAHE Enhanced")
plt.axis('off')
plt.show()
```

```python
gamma = 1.5  # >1 brightens, <1 darkens
gamma_corrected = np.array(255*(img / 255) ** gamma, dtype='uint8')

plt.figure(figsize=(6,6))
plt.imshow(gamma_corrected, cmap='gray')
plt.title("Gamma Corrected (gamma=1.5)")
plt.axis('off')
plt.show()
```

```python
# Using a simple kernel
kernel = np.array([[0, -1, 0],
                   [-1, 5,-1],
                   [0, -1, 0]])
sharpened = cv2.filter2D(img, -1, kernel)

plt.figure(figsize=(6,6))
plt.imshow(sharpened, cmap='gray')
plt.title("Sharpened Image")
plt.axis('off')
plt.show()
```

```python
smoothed = cv2.GaussianBlur(img, (5,5), 0)

plt.figure(figsize=(6,6))
plt.imshow(smoothed, cmap='gray')
plt.title("Smoothed Image")
plt.axis('off')
plt.show()
```
