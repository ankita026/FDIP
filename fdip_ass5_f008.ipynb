```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
from skimage import data

img = data.camera()
img = cv2.convertScaleAbs(img, alpha=0.5, beta=50)

plt.imshow(img, cmap='gray')
plt.title("Low Contrast Image")
plt.axis("off")
plt.show()
```

```python

min_val = np.min(img)
max_val = np.max(img)

contrast_stretched = ((img - min_val) / (max_val - min_val)) * 255
contrast_stretched = contrast_stretched.astype(np.uint8)

plt.imshow(contrast_stretched, cmap='gray')
plt.title("Contrast Stretched Image")
plt.axis("off")
plt.show()
```

```python
hist_eq = cv2.equalizeHist(img)

plt.imshow(hist_eq, cmap='gray')
plt.title("Histogram Equalized Image")
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(12,8))

plt.subplot(3,2,1), plt.imshow(img, cmap='gray'), plt.title("Low Contrast"), plt.axis("off")
plt.subplot(3,2,2), plt.hist(img.ravel(), 256, [0,256]), plt.title("Histogram (Low Contrast)")

plt.subplot(3,2,3), plt.imshow(contrast_stretched, cmap='gray'), plt.title("Contrast Stretched"), plt.axis("off")
plt.subplot(3,2,4), plt.hist(contrast_stretched.ravel(), 256, [0,256]), plt.title("Histogram (Stretched)")

plt.subplot(3,2,5), plt.imshow(hist_eq, cmap='gray'), plt.title("Histogram Equalized"), plt.axis("off")
plt.subplot(3,2,6), plt.hist(hist_eq.ravel(), 256, [0,256]), plt.title("Histogram (Equalized)")

plt.show()
```
