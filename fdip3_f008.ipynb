```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

```

```python

img = cv2.imread("0208.png")

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```python
def quantize(image, levels):
    factor = 256 // levels
    quantized = (image // factor) * factor
    return quantized
```

```python
q2 = quantize(gray, 2)
q4 = quantize(gray, 4)
q8 = quantize(gray, 8)

plt.figure(figsize=(12,4))
plt.subplot(1,4,1); plt.imshow(gray, cmap='gray'); plt.title("Original Gray"); plt.axis("off")
plt.subplot(1,4,2); plt.imshow(q2, cmap='gray'); plt.title("2 Levels"); plt.axis("off")
plt.subplot(1,4,3); plt.imshow(q4, cmap='gray'); plt.title("4 Levels"); plt.axis("off")
plt.subplot(1,4,4); plt.imshow(q8, cmap='gray'); plt.title("8 Levels"); plt.axis("off")
plt.show()
```
